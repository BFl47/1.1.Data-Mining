### Correlation Matrix with Target (Only relevant features)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a copied dataframe and select specific features
df_tmp = flight_train.copy()

selected_features = [
    'baseFare', 'total_distance', 'travelDuration_minutes', 'totalFlightTime', 
    'is_holiday', 'isAroundHoliday', 'num_legs', 'totalTransferTime', 'isNonStop', 'isCoach'
]

# Compute correlation matrix
corr_matrix = df_tmp[selected_features].corr()

# Plot heatmap
plt.figure(figsize=(20, 16))
heatmap = sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, vmin=-1, vmax=1, annot_kws={"size": 18})
plt.title("Feature Correlation Matrix With Target", fontsize=22)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
cbar = heatmap.collections[0].colorbar
cbar.ax.tick_params(labelsize=18)

plt.savefig("correlation_matrix_with_Target_updated2.png", dpi=300, bbox_inches="tight")
plt.show()
```
